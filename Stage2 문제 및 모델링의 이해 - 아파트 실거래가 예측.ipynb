í”„ë¡œì íŠ¸ì—ì„œëŠ” RMSE(
í‰ê·  ì œê³±ê·¼ ì˜¤ì°¨)ê°€ í‰ê°€ ê¸°ì¤€ìœ¼ë¡œ ì‚¬ìš©í•˜ê³  ìˆì–´ìš”.   
RMSE ì‹ì€ ë‹¤ìŒê³¼ ê°™ì´ MSE(í‰ê·  ì œê³± ì˜¤ì°¨)ì— ì œê³±ê·¼ì„ ì·¨í•œ í˜•íƒœì…ë‹ˆë‹¤! 
$${\displaystyle \operatorname {RMSE} ({\hat {\theta }})={\sqrt {\operatorname {MSE} ({\hat {\theta }})}}={\sqrt {\operatorname {E} (({\hat {\theta }}-\theta )^{2})}}.}$$

RMSEë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•œ mean_squared_error ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¨¼ì € ì„í¬íŠ¸í•´ì•¼ í•´ìš”!

```python
# RSME ì„±ëŠ¥í‰ê°€ ê³„ì‚°ì„ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬
from sklearn.metrics import mean_squared_error
```

mean_squared_errorë¡œ RMSEë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ëŠ” MSEì— ì œê³±ê·¼ì„ ì·¨í•˜ì—¬ ì•„ë˜ì™€ ê°™ì´ ì‘ì„±ë  ìˆ˜ ìˆì–´ìš”!

```python
def RMSE(y, y_pred):
    rmse = mean_squared_error(y, y_pred) ** 0.5
    return rmse
```

ì´ì œ RMSE í•¨ìˆ˜ë¡œ ëª¨ë¸ì„ ì˜ˆì¸¡í•´ë´…ì‹œë‹¤!
